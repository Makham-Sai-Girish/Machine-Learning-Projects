```python
#As seen above we are getting max and min value of each column and their respective row details
mdet=pd.DataFrame({})
ldet=pd.DataFrame({})
def maxi(df):
    columns=df.columns
    for col in columns:
        maxcol=data[col].idxmax()
        m=data.iloc[maxcol]
        mdet[col]=m
    return mdet
    
def mini(df):
    columns=df.columns
    for col in columns:
        mincol=data[col].idxmin()
        l=data.iloc[mincol]
        ldet[col]=l
    return ldet
```

defined function to get the details of max and min values of each column and display its rows as a dataframe

```python
maxi(data)
```

Each diagonal element gives the max value of that column

```python
mini(data)
```

```python
#Finding VIF to decide wether to remove or not
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(d):
    vif=pd.DataFrame()
    vif['Variable']=d.columns
    vif['vif factor']=[variance_inflation_factor(d.values,i) for i in range(d.shape[1])]
    return (vif)
```

we defined a function to check vif

```python
#Checking vif to ind the multicolinearity
calc_vif(x)
```

vif is high in many columns. So let us drop one column which having high multi collinearity and less correlation with Win
ER and ERA almost conveys the same thing, we can have ERA and remove ER since less impact on W also ER is correlated to RA which are similar

```python
#dropping the column ER
x.drop('ER',axis=1,inplace=True)
```

```python
#Importing the necessary modules
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
#Finding the best random state
maxr2=0
mse=0
best_score=0
rs=0
lr=LinearRegression()
for i in range(5000):
    #splitting the model 80% for train and 20% for test  
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr.fit(x_train,y_train)
    score=lr.score(x_train,y_train)
    predl=lr.predict(x_test)
    r2=r2_score(y_test,predl)
    ms=mean_squared_error(y_test,predl)
    if maxr2<r2:
        maxr2=r2
        mse=ms
        best_score=score
        rs=i
print(f"Best random state is: {rs}\n where best score is {best_score} \n where r2 score is {maxr2}\n and mean squared error is {mse}  ")
        
```

we got best random state is at 1398 where r2 score is 0.97 -> 93% accurate with mean squared error of 3.96

```python
#Making the predicted values rounded
predicted=[]
for i in predl:
    predicted.append(round(i))
    
```

```python
#So let us split at that parameters
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1398)
```

splitted the data successfully into train and test at random_state 1398

```python
#Making the data frame with actual and predicted values
dfr=pd.DataFrame({})
dfr['Actual']=y_test
dfr['Predicted']=predicted
dfr
```

```python
#viewing the graph about performance of the model using distribution plot
ax1=sns.distplot(dfr['Predicted'],hist=False,color='r',label='Predicted')
sns.distplot(dfr['Actual'],hist=False,color='g',label='Actual',ax=ax1)
```

From the above graph we can see.the model is overfitted, i.e the model is not learning all while making the best fit. 

```python
#Let us import and try other models
from sklearn.linear_model import Lasso, Ridge
from sklearn.linear_model import ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
#Making the list of models 
models=[LinearRegression(),Lasso(),Ridge(),ElasticNet(),DecisionTreeRegressor(),KNeighborsRegressor(),SVR()]
for m in models:
    m.fit(x_train,y_train)
    predm=m.predict(x_test)
    print(f"score of {m}:",m.score(x_train,y_train))
    print(f'Mean Squared error:',mean_squared_error(y_test,predm))
    print(f'r2 score:',r2_score(y_test,predm))
    print('\n')
```

From above we can see Linear Regression has the best r2 score of 0.97 and score of 0.93, where decision tree has the score 1 but r2 score is 0.21 which is very less 

```python
#Let us check cross validation to figure out the best model
from sklearn.model_selection import cross_val_score,KFold
#kf=KFold(5,shuffle=True)
for m in models:
    cvscore=cross_val_score(m,x,y,cv=4)
    print(f"Mean cv score of {m}:",cvscore.mean())
    print(f"std for cv score of {m}:",cvscore.std())
    print("\n")
    
```

By comparing we can see that Ridge regression has the least difference b/w r2 score and CV score

```python
r=Ridge()
r.fit(x_train,y_train)
predr=r.predict(x_test)
print("Score of Ridge:",r.score(x_train,y_train))
print("mse:",mean_squared_error(y_test,predr))
print("r2 score:",r2_score(y_test,predr))
```

made a model of ridge regression r, we can see it's r2 score is 0.68 with mean squared error of 48.86

```python
#Hyper parameter tuning of Ridge regression
from sklearn.model_selection import GridSearchCV
alp=np.random.uniform(low=0, high=100, size=(50,))
par={'alpha':alp,'fit_intercept':[True,False],
         'normalize':[True,False],
         'solver':['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga', 'lbfgs'],
         'positive':[True,False]}
grid=GridSearchCV(r,param_grid=par)
grid.fit(x_train,y_train)
print("Best score:",grid.best_score_)
print("Best estimator:",grid.best_estimator_)
print("Best parameters:",grid.best_params_)
```

We got the best parameters.Now making model with those parameters

```python
r=Ridge(alpha=0.36,normalize=True,solver='saga',fit_intercept=True,positive=False)
r.fit(x_train,y_train)
predr=r.predict(x_test)
print("Score of Ridge:",r.score(x_train,y_train))
print("mse:",mean_squared_error(y_test,predr))
print("r2 score:",r2_score(y_test,predr))
```

From the parameters we can see the r2 score is now 0.81 and mse is 28.3 -> Improved than earlier (r2- 0.68 , mse- 48.86)

Trying to improve the accuracy further more by checking the parameter randomstate

```python
r2=0
mse=0
rs=0
for i in range(50001):
    r=Ridge(alpha=0.36,normalize=True,random_state=i,solver='saga',fit_intercept=True,positive=False)
    r.fit(x_train,y_train)
    predr=r.predict(x_test)
    r2s=r2_score(y_test,predr)
    mses=mean_squared_error(y_test,predr)
    if r2s>r2:
        r2=r2s
        mse=mses
        rs=i
        
print("Score of Ridge:",r.score(x_train,y_train))
print("mse:",mse)
print("r2 score:",r2)
print(i)
```

By passing the random state , the r2 score increased from 0.813 to 0.817 and mse reduced to 27.7 from 28.3
