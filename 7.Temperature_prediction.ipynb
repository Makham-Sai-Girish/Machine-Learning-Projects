```python
def check(x):
    x.fillna(np.nan)
    sk=x.skew()
    if abs(sk) > 0.5:
        x.replace(np.nan,x.median(),inplace=True)
    else:
        x.replace(np.nan,x.mean(),inplace=True)
    return
```

```python
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(x):
    vif=pd.DataFrame()
    vif['variables']=x.columns
    vif['vif factor']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return(vif)
```

```python
#Finding the best random state
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

lr=LinearRegression()
r2=0
rs=0
for i in range(1000):
    x_max_train,x_max_test,y_max_train,y_max_test=train_test_split(x_max_scaled,y_max,test_size=0.2,random_state=i)
    lr.fit(x_max_train,y_max_train)
    pred=lr.predict(x_max_test)
    r2s=r2_score(y_max_test,pred)
    if r2s>r2:
        r2=r2s
        rs=i
print(f'best r2 score:{r2}, at random state: {rs}')
        
```

```python
models=[LinearRegression(),RandomForestRegressor(),GradientBoostingRegressor(),KNeighborsRegressor(),SVR()]
for m in models:
    m.fit(x_max_train,y_max_train)
    predm=m.predict(x_max_test)
    print(f"metrics of {m}:")
    print('train score:',m.score(x_max_train,y_max_train))
    print('test score:',m.score(x_max_test,y_max_test))
    print('r2 score:',r2_score(y_max_test,predm))
    cvscore=cross_val_score(m,x_max_scaled,y_max,cv=5)
    print('mean cv score:',cvscore.mean())
    print('Difference:',abs(r2_score(y_max_test,predm)-cvscore.mean()))
    print('\n\n')
```

from above we can see Linear regression has the least diference between cv score and r2 score and accuracy is 76% prediction, with near good fit since train and test scores are near.Let us improve this model
